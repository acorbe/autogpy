```python
#package import
import autogpy
```

```python
#generating some data to plot via np
import numpy as np
xx = np.linspace(-10,10,100)
yy = np.sin(xx)
zz = np.cos(xx)
```

```python
#actual autogpy figure
#notice the file tree: only the notebook exists!

with autogpy.Figure('test-figure',
                   tikz_enabled = True) as f:
    f.plot(xx,yy,title='sin',w='lines')
    f.plot(xx,zz,title='cos',lt=7)
    f.set(xlabel__e='t',ylabel__e='f(t)')
    
#see file tree on the left.
# all the gnuplot source files + makefile and data
# have been generated.

# we can easily re-generate the figure.
```
