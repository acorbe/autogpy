```python
## some test data
xx = np.linspace(0,10,100)
yy = 2*xx + np.random.normal(scale=0.2,size=xx.size)
```

```python
with autogpy.Figure('linear-fit-example-output') as f:
    f.plot(xx,yy,t='data')
    f.fit("foo(x)=a*x+bq",xx,yy)
    f.plot("foo(x)",t='fit')
```
